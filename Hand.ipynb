```python
import cv2
import mediapipe as mp
import numpy as np

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

# Function to calculate angle between three points
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
        
    return angle

# Function to get label
def get_label(num, hand, results):
    # Get the right hand landmarks
    landmarks = results.multi_hand_landmarks[num].landmark
    
    # Calculate the centroid of the hand
    x = [landmark.x for landmark in landmarks]
    y = [landmark.y for landmark in landmarks]
    z = [landmark.z for landmark in landmarks]
    x_min, x_max = min(x), max(x)
    y_min, y_max = min(y), max(y)
    z_min, z_max = min(z), max(z)
    
    # Get the index finger and pinky finger tip landmarks
    index_tip = np.array([landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP.value].x,
                          landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP.value].y])
    pinky_tip = np.array([landmarks[mp_hands.HandLandmark.PINKY_TIP.value].x,
                          landmarks[mp_hands.HandLandmark.PINKY_TIP.value].y])
    
    # Calculate the distance between the index and pinky finger tips
    distance = np.linalg.norm(pinky_tip - index_tip)
    
    # Calculate the average distance between all finger tips
    finger_tips = np.array([[landmark.x, landmark.y] for landmark in landmarks 
                             if landmark.HasField('x') and landmark.HasField('y')])
    avg_distance = np.mean(np.linalg.norm(finger_tips - finger_tips[mp_hands.HandLandmark.INDEX_FINGER_TIP.value], axis=1))
    
    # Print average distance for debugging
    print("Average Distance:", avg_distance)
    
    # If the average distance is smaller than a threshold, consider the hand closed
    if avg_distance < 0.05:  # Adjust threshold for closed hand detection
        return "Closed", (int((x_min + x_max) * 320), int((y_min + y_max) * 240))
    else:
        return "Open", (int((x_min + x_max) * 320), int((y_min + y_max) * 240))

# Main code
cap = cv2.VideoCapture(0)

# Initialize MediaPipe Hands model
with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    counter = 0  # Initialize counter for open and close hand gestures
    prev_gesture = None  # Store previous hand gesture to detect changes
    
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            break

        # BGR 2 RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Flip on horizontal
        image = cv2.flip(image, 1)

        # Set flag
        image.flags.writeable = False

        # Detections
        results = hands.process(image)

        # Set flag to true
        image.flags.writeable = True

        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Rendering results
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS,
                                          mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                          mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2),
                                          )

                # Render left or right detection
                text, coord = get_label(num, hand, results)
                cv2.putText(image, text, coord, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
                
                # Increment counter if hand gesture changes from closed to open
                if prev_gesture == "Closed" and text == "Open":
                    counter += 1
                prev_gesture = text

        # Display the count of open and close hand gestures
        cv2.putText(image, f"Open/Close Count: {counter}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

        # Display the frame
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
#final output

import cv2
import mediapipe as mp
import numpy as np

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

# Function to calculate angle between three points
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360 - angle
        
    return angle

# Function to get label
def get_label(num, hand, results):
    # Get the right hand landmarks
    landmarks = results.multi_hand_landmarks[num].landmark
    
    # Calculate the centroid of the hand
    x = [landmark.x for landmark in landmarks]
    y = [landmark.y for landmark in landmarks]
    z = [landmark.z for landmark in landmarks]
    x_min, x_max = min(x), max(x)
    y_min, y_max = min(y), max(y)
    z_min, z_max = min(z), max(z)
    
    # Get the index finger and pinky finger tip landmarks
    index_tip = (landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP.value].x,
                 landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP.value].y)
    pinky_tip = (landmarks[mp_hands.HandLandmark.PINKY_TIP.value].x,
                 landmarks[mp_hands.HandLandmark.PINKY_TIP.value].y)
    
    # Calculate the distance between the index and pinky finger tips
    distance = np.sqrt((pinky_tip[0] - index_tip[0])**2 + (pinky_tip[1] - index_tip[1])**2)
    
    # Get the landmarks for the thumb, index, middle, ring, and pinky fingers
    thumb_tip = (landmarks[mp_hands.HandLandmark.THUMB_TIP.value].x,
                 landmarks[mp_hands.HandLandmark.THUMB_TIP.value].y)
    index_tip = (landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP.value].x,
                 landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP.value].y)
    middle_tip = (landmarks[mp_hands.HandLandmark.MIDDLE_FINGER_TIP.value].x,
                  landmarks[mp_hands.HandLandmark.MIDDLE_FINGER_TIP.value].y)
    ring_tip = (landmarks[mp_hands.HandLandmark.RING_FINGER_TIP.value].x,
                landmarks[mp_hands.HandLandmark.RING_FINGER_TIP.value].y)
    pinky_tip = (landmarks[mp_hands.HandLandmark.PINKY_TIP.value].x,
                 landmarks[mp_hands.HandLandmark.PINKY_TIP.value].y)
    
    # Calculate the angles between fingers
    thumb_index_angle = calculate_angle(thumb_tip, index_tip, middle_tip)
    index_middle_angle = calculate_angle(index_tip, middle_tip, ring_tip)
    middle_ring_angle = calculate_angle(middle_tip, ring_tip, pinky_tip)
    
    # If the hand is fully open (all angles are above a certain threshold), consider it completely open
    if thumb_index_angle > 160 and index_middle_angle > 160 and middle_ring_angle > 160:
        return "Completely Open", (int((x_min + x_max) * 320), int((y_min + y_max) * 240))
    # If the distance is smaller than a threshold, consider the hand closed
    elif distance < 0.08:
        return "Closed", (int((x_min + x_max) * 320), int((y_min + y_max) * 240))
    # Otherwise, consider it open
    else:
        return "Open", (int((x_min + x_max) * 320), int((y_min + y_max) * 240))

# Main code
cap = cv2.VideoCapture(0)

# Initialize MediaPipe Hands model
with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    counter = 0  # Initialize counter for open and close hand gestures
    prev_gesture = None  # Store previous hand gesture to detect changes
    
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            break

        # BGR 2 RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Flip on horizontal
        image = cv2.flip(image, 1)

        # Set flag
        image.flags.writeable = False

        # Detections
        results = hands.process(image)

        # Set flag to true
        image.flags.writeable = True

        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Rendering results
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS,
                                          mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                          mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2),
                                          )

                # Render left or right detection
                text, coord = get_label(num, hand, results)
                cv2.putText(image, text, coord, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
                
                # Increment counter if hand gesture changes from closed to open
                if prev_gesture == "Closed" and text == "Open":
                    counter += 1
                prev_gesture = text

        # Display the count of open and close hand gestures
        cv2.putText(image, f"Open/Close Count: {counter}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

        # Display the frame
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
